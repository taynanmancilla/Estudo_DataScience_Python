# 6. Variáveis Quantitativas e Qualitativas
Variáveis Quantitativas são as que podem ser expressas em números

E as Variaveis Qualitativas são as que podem, de certa maneira, agregar em categorias


```python
# Nesse exemplo, todas as variveis são quantitativas

lren.dtypes
```

```python
# Mudar o formato dos dados do dataframe

cotacoes_ativos_long = pd.melt(lren, ignore_index=False) # pd.MELT: transforma tudo que é coluna, em linha | ignore_index=false: mantem extrutura do index(amplia o n° de linhas)

cotacoes_ativos_long
```

```python
# Tipo de dado 'object' nos remete as qualidade qualitativas (Conseguimos agrupar em categorias)

cotacoes_ativos_long.dtypes
```

```python
# Retorna as categorias unicas agrupadas do dataframe

set(cotacoes_ativos_long.variable)
```
